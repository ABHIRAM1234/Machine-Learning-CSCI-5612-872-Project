```python
MODIS_SP_DF=pd.DataFrame()
```

```python
from datetime import datetime, timedelta

start_date = datetime(2000, 11, 1)
end_date = datetime(2012, 1, 19)

current_date = start_date
while current_date <= end_date:
    dt=current_date.strftime('%Y-%m-%d')
    area_url = 'https://firms.modaps.eosdis.nasa.gov/api/area/csv/' + MAP_KEY + '/MODIS_SP/-125.0,24.396,-66.934,49.384/1/'+dt
    df_area = pd.read_csv(area_url)
    MODIS_SP_DF=pd.concat([MODIS_SP_DF,df_area],ignore_index=True)
    current_date += timedelta(days=1)
```
